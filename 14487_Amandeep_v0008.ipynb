```python
# ram ram

# input format

# john => 1st name
# doe => 2nd name
# 1 => for auto,for else type 0
# 5 => Position there symbol is to be filled
# and so on...

# For generating random number
import random as r

# Initializing a special class for all the functionalities and attributes reqed. for the game
# Used camelCase naming convention
class ticTacToe:
    # Constructor, called when object initialized
    def __init__(self):
        # Various attributes reqed. are declared here
        # Kept None for (little) memory efficiency
        self.oName = None
        self.xName = None
        self.flag = False
        self.board = {}
        self.filledPositions = []
        self.currentTurn = 'X'
        self.turnToName = {'X': '',
                           'O': '',
                           }
    
    # Tried keeping methods' name intuitive to make code more readable
    def makeBoard(self):
        # Used 1,2,3,... as str as later when feeding values to corresponding key, type casting won't be required
        # That is eventually lead to less error-prone code as user can type an alphabet instead of number
        self.board = {
            '1': '', '2': '', '3': '',
            '4': '', '5': '', '6': '',
            '7': '', '8': '', '9': '',
        }
        # Inserted '#' to have better readability when printing board for vacant spaces
        # Iterating over keys of dictionary 'board' and assigning values
        for keys in self.board:
            self.board[keys] = '#'
        print('\nBoard created!\n')
    
    # Formatting the dictionary into ticTacToe pattern
    def showBoard(self):
        # position are like the numpad on keyboard 
        print(self.board['7'] + '|' + self.board['8'] + '|' + self.board['9'])
        print('-+-+-')
        print(self.board['4'] + '|' + self.board['5'] + '|' + self.board['6'])
        print('-+-+-')
        print(self.board['1'] + '|' + self.board['2'] + '|' + self.board['3'])
        
    # Make sure names entered are at least satisfying the intention of assigning a name
    # For eq. length should be greater than 2 and names should not be same (at least in the game)
    # Used multiple conditions in one line using 'and' keyword i.e. all the conditions should be satisfied in order to move forward
    # Return True for acceptable names, false otherwise
    def areValidNames(self):
        if len(self.oName) > 2 and len(self.xName) > 2 and self.oName != self.xName:
            return True
        else:
            return False

    # Func. to get names of players along with names validation
    # Beautifully asks for names again, instead of just crashing or showing the exception
    def getPlayerNames(self):
        while not self.flag:
            self.oName = input('Enter the name for the player with \'O\' symbol:')
            self.xName = input('Enter the name for the player with \'X\' symbol:')
            self.flag = self.areValidNames()
            if not self.flag:
                print('\n--------ERROR---------\nNames must be greater than 2 '
                      'alphabets/numbers.\n----------------------\n')
            else:
                # turnToName is a dict with keys as symbols and values as corresponding names
                # Makes it easier to fetch name of symbol turn
                self.turnToName['X'] = self.xName
                self.turnToName['O'] = self.oName
                if self.oName[0].isalpha():
                    self.oName = self.oName[0].upper() + self.oName[1:]
                if self.xName[0].isalpha():
                    self.xName = self.xName[0].upper() + self.xName[1:]
                # Cheering them up for a simple yet sophisticated ride!
                print(f'\nOkay {self.oName} and {self.xName}, let\'s get ready for the game!\n')

    # When one player completes its turn, this func. is called 
    # Changes turn from x to o and vice-versa
    # returns changed turn
    def changeTurn(self):
        if self.currentTurn == 'X':
            self.currentTurn = 'O'
        else:
            self.currentTurn = 'X'
        return self.currentTurn

    # Important thingy! 
    # filledPositions list keeps track of all the positions that have been assigned some symbol
    # If I want to check for a position, I will simple use 'in' keyword. Easy peasy
    # returns default value i.e. None
    def lockPosition(self, position):
        self.filledPositions.append(position)

    # Checks if the position user enters, irrespective of the symbol, is empty
    # returns true if position filled, false otherwise
    def isPositionFilled(self, position):
        if position in self.filledPositions:
            return True
        else:
            return False

    # Majestic function to input the symbols into board (or dictionary)
    # Especially for user input as that is more prone to errors
    # More information inside the function block/body
    def inputIntoBoard(self):
        # Loop to get get position again if user gives unacceptable input
        while True:
            # Notifies turn of user with its name and symbol
            position = input(f'Turn of user: {self.turnToName[self.currentTurn]}'
                             f' bearing symbol: {self.currentTurn}: ')
        
            # In the middle of 'battlefield', if user has to leave the match
            # He/She can easily show white flag to opponent declaring his win and this one's, lose
            if position == 'exit' or position == 'Exit' or position == 'EXIT':
                print(f'User {self.turnToName[self.currentTurn]}, you have chosen to quit in between. This is turn '
                      f'would make other user a WINNER!')
                return False

            # Position validation we talked about earlier
            # returns True of inserted successfully else reruns to get input again
            if not self.isPositionFilled(position) and 0 < int(position) < 10:
                self.lockPosition(position)
                # Assigning symbol i.e. value to input position i.e. key
                self.board[position] = self.currentTurn
                print(f'Inserted successfully at position {position} for symbol {self.currentTurn}\n\n')
                return True
            else:
                print('\n--------ERROR---------\nPosition is already filled or '
                      'greater than required value. Try somewhere else.\n----------------------\n')
                print(self.showBoard())

    # Specially for bot input
    # Minimal error prone %age
    # hence, not much exception handling required here
    # could have integrated this into inputIntoBoard(), but that would have made that module very complicated
    # and hence, difficult to optimize, upgrade or modify
    # return None (default)
    def botInput(self):
        for _ in range(9):
            # Generating random int b/w 1-9
            random = r.randint(1, 9)
            # Type casting to str as dictionary has keys as str
            # Type casting to str is less error prone, so acceptable
            random = str(random)
            # Obvious validations we talked about earlier
            if not self.isPositionFilled(random) and 0 < int(random) < 10:
                self.lockPosition(random)
                self.board[str(random)] = self.currentTurn
                print(f'Computer named {self.turnToName[self.currentTurn]} has inserted {self.currentTurn} to position'
                      f' {random}.')
                print(self.showBoard())
                break

    # Checking for win
    def checkWin(self, turn):
        # couldnt wrap this text as it was leasding to some logical error
        if self.board['7'] == self.board['8'] == self.board['9'] != '#' or self.board['4'] == self.board['5'] == self.board['6'] != '#' or self.board['1'] == self.board['2'] == self.board['3'] != '#' or self.board['7'] == self.board['4'] == self.board['1'] != '#' or self.board['8'] == self.board['5'] == self.board['2'] != '#' or self.board['9'] == self.board['6'] == self.board['3'] != '#' or self.board['1'] == self.board['5'] == self.board['9'] != '#' or self.board['3'] == self.board['5'] == self.board['7'] != '#':
            return f'Congratulations user {self.turnToName[self.currentTurn]} bearing symbol {self.currentTurn}. You ' \
                   f'have won the game!'
        # for turn == 8, all turns used adn hence no winner as above condition no satisfied
        elif turn == 8:
            return f'It is a DRAW. Well played {self.xName} and {self.oName}'
        else:
            pass


flag = True
# loop to amke driver code run again if user wants to play again
while flag:
    startGame = ticTacToe() # declaring object of class ticTacToe()
    startGame.makeBoard() # called function of startGame object
    startGame.getPlayerNames() # and called this too and so on...
    isAuto = int(input('Auto: 1 and Manual: 0  '))
    # for manual mode
    if isAuto == 0:
        for inr in range(9):
            x = startGame.inputIntoBoard()
            print(startGame.showBoard())
            if startGame.checkWin(inr) is not None and inr > 2:
                print(startGame.checkWin(inr))
                break
            if not x:
                print('Exiting')
                break
            startGame.changeTurn()
    
    # for auto mode
    if isAuto == 1:
        for itr in range(9):
            if itr % 2 != 0:
                startGame.botInput()
            else:
                x = startGame.inputIntoBoard()
            if startGame.checkWin(itr) is not None and itr > 2:
                print(startGame.checkWin(itr))
                break
            if not x:
                print('Exiting')
                break
            startGame.changeTurn()
    
    # asks to play agian or not and acts accordingly
    print('Wanna play again? (y/n)')
    if input() == 'n':
        print('Goodbye. Hope you\'ve enjoyed this little game.')
        flag = False
        
        
# Thankyou for checking this notebook.
# Tried to keep intuitive name for better unserstanding
# used classe because: -
# more readability
# more maintainability
# organized as attributes and functions are grouped in one class
# easy to maintain code
# more structured
# easy for any upgradation
# used camelCase naming convention for better naming instead of underscores (atleast for my use-case)
# code reusability => changeTurns() for changing turns instead of doing again and again by writing code, simply call this function
# and much more
```
